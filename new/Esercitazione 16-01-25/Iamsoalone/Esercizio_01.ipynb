```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

#Punto a
def simulazione_lancio(p, N):
    print("Simulazione di lancio moneta con probabilità p =", p, "per", N, "occorrenze")

    #Generiamo le prove in base al numero in input
    prove = np.random.rand(N)
    print("Prove generate [0,1] (", N, ")")
    print(prove)
    print(" ")

    #Occorrenze di testa per il tentativo (indicato dalla cella dell'array)
    testa = np.where(prove < p, 1, 0)
    print("Occorrenze di testa / croce per ogni numero di tentativo")
    print(testa)
    print(" ")

    #Verifica delle frequenze
    frequenze = np.unique(testa, return_counts = True)
    print("Frequenze per valori")
    print("Occorrenze di croce (0):", frequenze[1][0])
    print("Occorrenze di testa (1):", frequenze[1][1])
    print(" ")

    ##Implementazione per il plot del punto b
    plt.bar(frequenze[0], frequenze[1], width = 0.5, align = 'center')
    plt.xticks([0, 1])
    plt.axhline(N*p,frequenze[0][0], frequenze[0][1], color = 'r')
    title = "N = " + str(N)
    plt.title(title)
    plt.xlabel("Testa - Croce")
    plt.ylabel("Frequenze")
    plt.show()
    print("Osservazione: data la probabilità p =", p, "Il nostro threshold è la frequenza", N*p)
```

```python
N = 100
p = 0.5
simulazione_lancio(p, N)
```

```python
N = 1000
p = 0.5
simulazione_lancio(p, N)
```

```python
N = 10000
p = 0.5
simulazione_lancio(p, N)
```
