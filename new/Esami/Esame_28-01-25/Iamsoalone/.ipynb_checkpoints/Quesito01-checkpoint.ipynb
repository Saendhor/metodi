```python
import numpy as np

x_values = np.array([1213, 1518, 1550, 1215, 2120, 2207])
y_values = np.array([15474, 16497, 17224, 14459, 22186, 23483])
```

```python
# Punto a) Stabilire se tra i dati X e dati Y c'è correllazione
n = x_values.size
x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean)*(y_values - y_mean))
sigma_x_quadro = (1/n) * np.sum((x_values - x_mean)**2)
sigma_y_quadro = (1/n) * np.sum((y_values - y_mean)**2)

print("I valori di sigma: xy, x, y", sigma_xy, sigma_x_quadro, sigma_y_quadro)

rho = sigma_xy / np.sqrt((sigma_x_quadro)*(sigma_y_quadro))
print("Il valore di rho è:", rho)

'''
COMMENTO
Essendo rho prossimo ad uno vuol dire che esiste una correlazione forte fra le v.a. X e Y
'''
```

```python
#Punto b) Usare il metodo dei minimi quadrati per determinare la retta di regressione.
# In un plot mostrare i dati e la retta ricavata
import matplotlib.pyplot as plt

b1 = sigma_xy / sigma_x_quadro
b0 = y_mean - (b1*x_mean)

print("b0, b1", b0, b1)

x_graph = np.linspace(min(x_values), max(x_values), 10000)
y_graph = b0 + (b1 * x_graph)

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_values, y_values, 'ro', label = "Punti originali")
plt.xlabel("Valori in x compresi fra min e max di X")
plt.ylabel("Valori ricavati sulla base di x_graph")
plt.legend()
plt.show()
```

```python
#Punto c) Calcolare il coefficiente di determinazione e commentare il risultato

R = (sigma_xy**2)/ (sigma_x_quadro * sigma_y_quadro)
print("Il coefficiente di determinazione R è", R)

'''
COMMENTO
Essendo il coefficiente di determinazione R prossimo a 1 vuol dire che il modello di regressione adottato è appropriato
'''
```

```python
#Punto d) Determinare la nuova retta di regressione e confrontarla con la precedente.

nx_values = np.array([1213, 1518, 1550, 1215, 2120, 2207, 2175, 1128])
ny_values = np.array([15474, 16497, 17224, 14459, 22186, 23483, 24095, 13982])

nn = np.size(nx_values)
nx_mean = np.mean(nx_values)
ny_mean = np.mean(ny_values)

nsigma_xy = (1/nn) * np.sum((nx_values - nx_mean)*(ny_values - ny_mean))
nsigma_x_quadro = (1/nn) * np.sum((nx_values - nx_mean)**2)
nsigma_y_quadro = (1/nn) * np.sum((ny_values - ny_mean)**2)
print(nsigma_xy, nsigma_x_quadro, nsigma_y_quadro)

nb1 = nsigma_xy/nsigma_x_quadro
nb0 = ny_mean - (nb1 * nx_mean)
print(nb0, nb1)

nx_graph = np.linspace(min(nx_values), max(nx_values), 10000)
ny_graph = nb0 + (nb1 * nx_graph)
plt.plot(nx_graph, ny_graph, label = "Retta di regressione (2)")
plt.plot(nx_values, ny_values, 'ro', label = "Punti originali (2)")
plt.xlabel("Valori in nx compresi fra min e max di nX")
plt.ylabel("Valori ricavati sulla base di nx_graph")
plt.legend()
plt.show()

nrho = nsigma_xy / np.sqrt(nsigma_x_quadro * nsigma_y_quadro)
print("Il valore di nrho è", nrho)

'''
COMMENTO
Il valore di rho prima era 0.988 8448849296708
L'attuale valore di rho, invece, è nrho = 0.98 87072074302627

Chiaramente non stiamo parlando di differenze significative dal momento che aggiungiamo solo 2 valori, ma il secondo modello di regressione sembra essere più accurato in corrispondenza dei dati rispetto quello precedentemente calcolato.
'''
```
