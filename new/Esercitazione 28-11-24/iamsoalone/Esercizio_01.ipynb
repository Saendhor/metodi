```python
import numpy as np
x_values = np.array([211, 332, 322, 435, 275])
y_values = np.array([112, 155, 131, 208, 138])
```

```python
# 1 Si calcoli il coefficiente di correlazione fra le variabili x e y
n = x_values.size

x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean) * (y_values - y_mean))

sigma_quadro_x = np.var(x_values, mean = x_mean)
sigma_quadro_y = np.var(y_values, mean = y_mean)

dummytest = (1/n) * np.sum((x_values - x_mean)**2)
print(dummytest)
print("I valori di sigma sono (xy, x, y): ", sigma_xy, sigma_quadro_x, sigma_quadro_y)


rho = sigma_xy / np.sqrt(sigma_quadro_x * sigma_quadro_y)
print(rho)

# sup - min
if np.abs(rho) <= 1 and np.abs(rho) >= 0.7:
    print("Correlazione forte", np.abs(rho))
elif np.abs(rho) <= 0.7 and np.abs(rho) >= 0.3:
    print("Correlazione moderata", np.abs(rho))
elif np.abs(rho) <= 0.3 and np.abs(rho) >= 0:
    print("Correlazione debole", np.abs(rho))

#Siccome rho è prossimo ad 1, possiamo procedere col punto 2
```

```python
#2 Usare il metodo dei minimi quadrati per determinare la retta di regressione.
#  Fare un plot dei dati e della retta di regressione
import matplotlib.pyplot as plt

b1 = sigma_xy / sigma_quadro_x
b0 = y_mean - (b1 * x_mean)
print("La retta di regressione è y =", b0, "+", b1, "x")

x_graph = np.linspace (min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)
plt.title("Rappresentazione tempo di calcolo / dati")
plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_values, y_values, '*', label = "Valori originali")
plt.xlabel("Numero di dati")
plt.ylabel("Tempo impiegato a processare [s]")
plt.legend()
plt.show()
```

```python
#3 Si calcoli il tempo previsto per processare 200, 300, 400, 500 dati
print("Il risultato previsto per processare 200 dati è", round(b0 + (b1*200), 3),"secondi")
print("Il risultato previsto per processare 300 dati è", round(b0 + (b1*300), 3),"secondi")
print("Il risultato previsto per processare 400 dati è", round(b0 + (b1*400), 3),"secondi")
print("Il risultato previsto per processare 500 dati è", round(b0 + (b1*500), 3),"secondi")
```
