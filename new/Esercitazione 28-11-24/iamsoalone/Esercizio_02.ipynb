```python
import numpy as np
x_values = np.array([0.3, 0.4, 0.2, 0.5, 0.6])
y_values = np.array([0.11, 0.14, 0.06, 0.16, 0.22])
```

```python
#01 - Coefficiente correlazione rho
n = x_values.size
x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean) * (y_values - y_mean))
sigma_quadro_x = np.var(x_values, mean = x_mean)
sigma_quadro_y = np.var(y_values, mean = y_mean)

rho = sigma_xy / np.sqrt(sigma_quadro_x * sigma_quadro_y)
print("Il valore di rho è", rho)
```

```python
#02 - Ragionevole supporre ci sia una relazione lineare fra le due quantità?
## COMMENTO ##
# Introduciamo la variabile rho al fine di comprendere in che tipo di correlazione stanno le due variabili.
#In particolare:
    #se 1 >= rho >= 0.7 -> correlazione forte
    #se 0.7 >= rho >= 0.3 -> correlazione moderata
    #se 0.3 >= rho >= 0 -> correlazione debole
# Il valore calcolato di rho è pari a 0.98~, quindi possiamo dire che esiste una correlazione fra x e y e nello specifico è una correlazione forte.
```

```python
#03 - Retta di regressione + plot dati e retta
import matplotlib.pyplot as plt
b1 = sigma_xy / sigma_quadro_x
b0 = y_mean - (b1 * x_mean)
print ("I valori di b0 e b1 sono", b0, b1)

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_values, y_values, '*', label = "Valori originali")
plt.legend()
plt.show()
```

```python
#04 - Previsione più affidabile
x1 = 0.35
x2 = 0.8

y1 = b0 + (b1 * x1)
y2 = b0 + (b1 * x2)

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x1, y1, '*', label = "x1")
plt.plot(x2, y2, "o", label = "x2")

plt.legend()
plt.show()
#ritengo sia più affidabile il punto x1 dal momento che è sicuramente un punto coinvolto nella retta precedentemente calcolata.
```
