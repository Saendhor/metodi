```python
import numpy as np

x_values = np.array([211, 332, 322, 435, 275])
y_values = np.array([112, 155, 131, 208, 138])
```

```python
# Punto a) coefficiente di correlazione di X e Y
x_mean = np.mean(x_values)
y_mean = np.mean(y_values)
n = np.size(x_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean) * (y_values - y_mean))
sigma_x_quadro = (1/n) * np.sum((x_values - x_mean)**2)
sigma_y_quadro = (1/n) * np.sum((y_values - y_mean)**2)
print("sigma_xy, sigma²x, sigma²y: (",
      sigma_xy, sigma_x_quadro, sigma_y_quadro,")")

print(" ")
rho = sigma_xy / np.sqrt(sigma_x_quadro * sigma_y_quadro)
print("Il valore di rho è", rho)
```

```python
# Punto b) Retta di regressione + plot dati
import matplotlib.pyplot as plt
b1 = sigma_xy / sigma_x_quadro
b0 = y_mean - (b1 * x_mean)

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_values, y_values, 'ro', label = "Punti originali")
plt.show()
```

```python
#Punto c) tempo previsto per processare 200, 300, 400, 500 dati.
r1 = b0 + (b1 * 200)
r2 = b0 + (b1 * 300)
r3 = b0 + (b1 * 400)
r4 = b0 + (b1 * 500)
print("Tempo stimato:",
     r1, r2, r3, r4)
```
