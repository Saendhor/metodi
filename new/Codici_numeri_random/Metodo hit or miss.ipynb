Calcolo area integrale con il metodo hit or miss

```python
import numpy as np
from scipy.stats import norm 
import matplotlib.pyplot as plt

def gauss(x):
    return np.exp(-0.5*(x**2))/np.sqrt(2*np.pi) 

a = -3
b = 3
M = 1/np.sqrt(2*np.pi)

nmin = 100
nmax = 1000000
n = nmax-nmin 
nrange = np.arange(nmin, nmax, 1)
nsize = np.size(nrange)

x1 = np.random.rand(n)
csi = a + (b-a)*x1
x2 = np.random.rand(n)
eta = M*x2

y = np.where(eta <= gauss(csi), 1, 0)

Ns = np.cumsum(y)
I = M*Ns*(b-a)/nrange


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 7))
ax.plot(nrange, I, 'bo')
ax.plot(nrange, np.ones(nsize)*(norm.cdf(b)-norm.cdf(a)), 'r-')
ax.set_ylim([norm.cdf(b)-norm.cdf(a) - 0.1, norm.cdf(b)-norm.cdf(a) + 0.1])
plt.show()
```

Calcoliamo il pi greco

```python
M = 1
a = 0
b = 1

nmin = 100
nmax = 1000000
n = nmax-nmin 
nrange = np.arange(nmin, nmax, 1)
nsize = np.size(nrange)

x1 = np.random.rand(n)
csi = a + (b-a)*x1
x2 = np.random.rand(n)
eta = M*x2
y = np.where(eta*eta + csi*csi <= 1, 1, 0)
Ns = np.cumsum(y)
I = M*Ns*(b-a)/nrange #area quarto di cerchio

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 7))
ax.plot(nrange, 4*I, 'bo')
ax.plot(nrange, np.ones(nsize)*np.pi, 'r-')
ax.set_ylim(np.pi-0.01, np.pi+0.01)
plt.show()
```
