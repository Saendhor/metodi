```python
n = 10
p = 0.5
y = np.zeros(n)
for i in range(n):
    y[i] = binom.pmf(i, n, p)

plt.plot(range(n), y, '^', color = 'red')
```

```python
n = 100
p = 0.1
k = range(n)
y = binom.pmf(k, n, p)
plt.plot(k, y, '^', color = 'red')
plt.xlabel("numero di successi")
plt.ylabel("probabilità di avere quel numero di successi")
```

```python
n_p = 100
p_p = 0.1
la = n_p*p_p
k_p = range(n_p)
y_p = poisson.pmf(k_p, la)
plt.plot(k_p, y_p, 'o', color = 'blue')
```

```python
n = 1000
p = 0.1
la = n*p
k = range(n)
y = binom.pmf(k, n, p)
z = poisson.pmf(k,la)
plt.plot(k, y, 'r^', k, z, 'bo')
```

```python
#Calcoliamo la probabilità di ottenere la combinazione [0, 3, 5, 4, 0, 1] in un dado truccato
x = [0, 1, 5, 3, 0, 1]
p = [2/15, 2/15, 1/3, 2/15, 2/15, 2/15]
n = sum(x)
ris = multinomial.pmf(x, n, p)
print(ris)
```
