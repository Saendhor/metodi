```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon
```

```python
#Impariamo a usare la funzione uniform. 
#Come argomenti accetta il valore di partenza dell'intervallo e la sua ampiezza (qui la chiamiamo loc)
a = -5 
b = 5
loc = b-a
x = np.linspace(-7, 7, 100)
plt.plot(x, uniform.pdf(x, a, loc))
plt.show()
```

```python
#Plottiamo la funzione di ripartizione e la confrontiamo con quella derivata a lezione
plt.plot(x, uniform.cdf(x, a, loc))
plt.plot(x, (x-a)/(b-a), '--')
plt.show()
```

```python
#Confronto tra Gaussiane
#La funzione norm accetta due parametri, media e varianza. Se omessi genera normali standard.
mu = 0
mu1 = 1
var = 1
var1 = 2
var2 = 0.5
x = np.linspace(-7, 7, 100)
plt.plot(x, norm.pdf(x, mu, var), 'b')
#plt.plot(x, norm.pdf(x, mu, var1), 'r')
#plt.plot(x, norm.pdf(x, mu, var2), 'g')
plt.plot(x, norm.pdf(x, mu1, var))
plt.show()
```

```python
#Calcolo i quantili e coloro l'area sottesa fino a un determinato quantile
q = 0
alpha = norm.cdf(q, mu, var)
print(alpha)
qalpha=norm.ppf(alpha, mu, var)
print(qalpha)
plt.plot(x, norm.pdf(x, mu, var), 'b')
plt.fill_between(x, norm.pdf(x, mu, var), where= (-7 < x)&(x < q), color='b', alpha=0.2)
plt.show()
```

```python
#Plotto la Gaussiana in un intervallo che sottenda lo 0.99998 dell'area
xmin = norm.ppf(0.0001, mu, var)
xmax = norm.ppf(0.9999, mu, var)
print(xmin)
print(xmax)
x2 = np.linspace(xmin, xmax, 100)
plt.plot(x2, norm.pdf(x2, mu, var))
plt.fill_between(x2, norm.pdf(x2, mu, var), where= (-3 < x2)&(x2 < 3), color='b', alpha=0.2)
plt.show()
```

```python
#Verifico la regola 68-95-99.7
print(norm.cdf(1, mu, var)-norm.cdf(-1, mu,var))
plt.plot(x2, norm.pdf(x2, mu, var))
plt.fill_between(x2, norm.pdf(x2, mu, var), where= (-1 < x2)&(x2 < 1), color='b', alpha=0.2)
plt.show()
```

```python
print(norm.cdf(2, mu, var)-norm.cdf(-2, mu,var))
plt.plot(x2, norm.pdf(x2, mu, var))
plt.fill_between(x2, norm.pdf(x2, mu, var), where= (-2 < x2)&(x2 < 2), color='b', alpha=0.2)
plt.show()
```

```python
#Plotto la densità esponenziale
plt.plot(x, expon.pdf(x, 0, 1))
plt.show()
```

```python
#Calcolo e confronto i momenti della distribuzione Normale, uniforme e esponenziale
mn = norm.stats(mu, var, 'mvsk')
un = uniform.stats(a, loc, 'mvsk')
en = expon.stats(0, 1, 'mvsk')
print(mn)
print(un)
print(en)
```

m --> media
v --> varianza
s --> skewness
k --> kurtosis

s e k sono dei valori che indicano asimmetria e piattezza di una curva rispetto alla normale, che viene presa come riferimento (e infatti s = 0 e k = 0)

Se s è < 0 la curva è asimmetrica a sinistra, se s > 0 è asimmetrica a destra. 
Se k < 0 la curva è più larga di una normale (distribuzione uniforme), se k > 0 è più appuntita (distribuzione esponenziale)
