```python
import numpy as np
x_values = np.array([20, 25, 30, 35, 40, 50, 60, 65, 70, 75, 80, 90])
y_values = np.array([1.81, 1.70, 1.65, 1.55, 1.48, 1.40, 1.30, 1.26, 1.24, 1.21, 1.20, 1.18])

n = x_values.size
```

```python
# Punto 1) coefficienti della retta di regressione e i loro intervalli di confidenza al 95%. coefficiente di determinazione

x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean)*(y_values - y_mean))
sigma_quadro_x = (1/n) * np.sum((x_values - x_mean)**2)

b1 = sigma_xy / sigma_quadro_x
b0 = y_mean - (b1 * x_mean)
print("b0:", round(b0, 2), ", b1:", round(b1, 2))

from scipy.stats import t
alpha1 = 0.05

T = t.ppf(1- alpha1/2, n-2)

y_hat = b0 + (b1 * x_values)
r = y_values - y_hat
s_quadro = (1/(n-2)) * np.sum((r)**2)
s = np.sqrt(s_quadro)

infb0 = b0 - (s * T * np.sqrt((1/n) + ((x_mean**2) / n * sigma_quadro_x)))
supb0 = b0 + (s * T * np.sqrt((1/n) + ((x_mean**2) / n * sigma_quadro_x)))
print("Intervallo confidenza b0: [", round(infb0, 2), ",", round(supb0, 2), "]")

infb1 = b1 - (T * (s / np.sqrt(n * sigma_quadro_x)))
supb1 = b1 + (T * (s / np.sqrt(n * sigma_quadro_x)))
print("Intervallo confidenza b1: [", round(infb1, 2), ",", round(supb1, 2), "]")



sigma_quadro_r = (1/n) * np.sum(r**2)
sigma_quadro_f = (sigma_xy**2) / sigma_quadro_x
sigma_quadro_y = sigma_quadro_f + sigma_quadro_r
R_quadro = 1 - (sigma_quadro_r / sigma_quadro_y)
print("Coefficiente di determinazione R²:", round(R_quadro, 2))
```

```python
# Punto 2) Grafico
import matplotlib.pyplot as plt

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)

plt.title("Variazione costo unitario per numero prezzi richiesti")
plt.xlabel("Numero pezzi richiesti")
plt.ylabel("Costo unitario")

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_values, y_values, '*', label = "Valori originali")

plt.legend()
plt.show()
```

```python
# Punto 3) Si determinino i coefficienti per un modello di regressione lineare del tipo y = a0 + a1x + a2x2 e si calcoli nuovamente il coefficiente di determinazione

e = np.ones(n)
X = np.zeros((n, 3))
X [:, 0] = e
X [:, 1] = x_values
X [:, 2] = (x_values)**2

XX = np.linalg.pinv(X)
b = np.dot(XX, y_values)

y_hat2 = np.dot(X, b)
R_quadro2 = np.sum((y_hat2 - y_mean)**2) / np.sum((y_values - y_mean)**2)
print("Coefficiente di determinazione R²:", round(R_quadro2, 2))

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)

x_graph2 = np.linspace(min(x_values), max(x_values), 1000)
y_graph2 = b[0] + (b[1] * x_graph2) + (b[2] * (x_graph2**2))

plt.title("Variazione costo unitario per numero prezzi richiesti")
plt.xlabel("Numero pezzi richiesti")
plt.ylabel("Costo unitario")

plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.plot(x_graph2, y_graph2, label = "Modello quadratico di regressione")
plt.plot(x_values, y_values, '*', label = "Valori originali")

plt.legend()
plt.show()
```

```python
# Punto 4) Si determinino gli intervalli di confidenza al 95% per i nuovi coefficienti.
alpha4 = 0.05

k = 3
r2 = y_values - y_hat2
s_quadro2 = (1 / (n - k)) * np.sum(r2**2)
s2 = np.sqrt(s_quadro2)

T4 = t.ppf(1-alpha4, n-k)
M = np.linalg.inv(np.dot(X.T,X))

print(b[0], b[1], b[2])

infb0_2 = b[0] - (s2 * T4 * np.sqrt(M[0][0]))
supb0_2 = b[0] + (s2 * T4 * np.sqrt(M[0][0]))
print("Intervallo confidenza b[0]: [", round(infb0_2, 2), ",", round(supb0_2, 2), "]")

infb1_2 = b[1] - (s2 * T4 * np.sqrt(M[1][1]))
supb1_2 = b[1] + (s2 * T4 * np.sqrt(M[1][1]))
print("Intervallo confidenza b[1]: [", round(infb1_2, 2), ",", round(supb1_2, 2), "]")

infb2_2 = b[2] - (s2 * T4 * np.sqrt(M[2][2]))
supb2_2 = b[2] + (s2 * T4 * np.sqrt(M[2][2]))
print("Intervallo confidenza b[2]: [", round(infb2_2, 2), ",", round(supb2_2, 2), "]")
```
