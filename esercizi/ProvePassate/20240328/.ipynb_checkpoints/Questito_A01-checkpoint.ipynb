```python
from scipy.stats import norm
mi = 220 #g
devStandard = 8 #g
```

```python
# Punto 1) Qual è la probabilità che la confezione pesi più di 230 g?
x1 = 230
p1 = norm.cdf(x1, mi, devStandard)
q1 = 1 - p1
print("la probabilità che la confezione pesi più di 230 g è", round(q1, 2))
```

```python
# Punto 2) Qual `e la probabilità che il peso della confezione sia compreso tra 210 g e 225 g?

x2 = 210
y2 = 225

p2 = norm.cdf(y2, mi, devStandard) - norm.cdf(x2, mi, devStandard)
print("la probabilità che il peso della confezione sia compreso tra 210 g e 225 g è", round(p2, 2))
```

```python
# Punto 3) Le confezioni vengono imballate in scatole da 10 unità. Dopo aver calcolato la distribuzione del peso della scatola, calcolare la probabilit`a che il peso totale superi i 2250 g (si trascuri il peso della scatola).
import numpy as np
import matplotlib.pyplot as plt

x_graph = np.linspace(mi - (3*devStandard), mi + (3*devStandard))
y_graph = norm.pdf(x_graph, mi, devStandard)
plt.title ("Distribuzione peso nella scatola")
plt.plot (x_graph, y_graph)
plt.show()

n = 10
x3 = 2250

p3 = norm.cdf(x3, mi*n, devStandard * np.sqrt(n))
q3 = 1 - p3
print("la probabilità che il peso totale superi i 2250 g è", round(q3, 2))
```
