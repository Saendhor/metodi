```python
def f(x):
    a = -(3/4)
    b = 3/2
    return a * (x**2) + b * x
```

```python
# Algoritmo per generare numeri pseudo-casuali con distribuzione X utilizzando il metodo del rigetto.
import numpy as np

def rigetto (a, b ,M):
    while True:
        X = np.random.rand()
        xi = a + (b-a)*X
        eta = M*X
        if eta <= f(xi):
            break
    return xi
```

```python
import matplotlib.pyplot as plt

n = 100000

a = 0
b = 2
M = f(2)

X = np.zeros(n)
for i in range (n):
    X[i] = rigetto(a, b, M)

plt.hist(X, density = True)
x_graph = np.linspace(a, b, n)
y_graph = f(x_graph)
plt.plot(x_graph, y_graph)
plt.title("Distribuzione di probabilità di f")
plt.xlabel("Numeri nel range [0, 2]")
plt.ylabel("Risultato funzione nel range [0, 2]")
plt.show()
    
```
