```python
import numpy as np
y_values = np.array([2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]) #grammi solidi rimossi
x_values = np.array([4.3, 1.5, 1.8, 4.9, 4.2, 4.8, 5.8, 6.2, 7.0, 7.9]) #tempo asciugature [h]
n = x_values.size # dovrebbe essere 10
```

```python
# Punto 1) b0, b1; intervalli di confidenza al 95%. Coefficiente di determinazione
from scipy.stats import t
alpha = 0.05

x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean)*(y_values - y_mean))
sigma_quadro_x = (1/n) * np.sum((x_values - x_mean)**2)

b1 = sigma_xy / sigma_quadro_x
b0 = y_mean - (b1 * x_mean)

print ("I coefficienti hanno valore (b0, b1) =", "(", round(b0,2), ",", round(b1,2) ,")")

T = t.ppf(1 - alpha/2, n-2)

y_hat = b0 + (b1*x_values)
r = y_values - y_hat

s_quadro = (1/(n-2)) * np.sum((r)**2)
s = np.sqrt(s_quadro)

infb0 = b0 - (s * T * np.sqrt((1/n) + ((x_mean**2) / (n * sigma_quadro_x))))
supb0 = b0 + (s * T * np.sqrt((1/n) + ((x_mean**2) / (n * sigma_quadro_x))))

infb1 = b1 - (T * (s / np.sqrt(sigma_quadro_x * n)))
supb1 = b1 + (T * (s / np.sqrt(sigma_quadro_x * n)))

print("b0 compreso fra [", round(infb0, 2), ",", round(supb0, 2), "]")
print("b1 compreso fra [", round(infb1, 2), ",", round(supb1, 2), "]")

sigma_quadro_r = (1/n) * np.sum(r**2)
sigma_quadro_f = (sigma_xy **2) / sigma_quadro_x
sigma_quadro_y = sigma_quadro_f + sigma_quadro_r

R_quadro = 1 - (sigma_quadro_r / sigma_quadro_y)
print("Coefficiente di determinazione R² =", round(R_quadro, 2))
```

```python
# Punto 2) Dati e retta di regressione nello stesso grafico
import matplotlib.pyplot as plt

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1*x_graph)
plt.plot(x_values, y_values, "*", label = "Dati riportati")
plt.plot(x_graph, y_graph, label = "Retta di regressione")
plt.legend()
plt.xlabel("Tempo asciugature [h]")
plt.ylabel("Grammi solidi rimossi [g]")
plt.show()
```

```python
# Punto 3) stima grammi di solidi rimossi a 4.25 ore
x_req = 4.25
stima = b0 + (b1 * x_req)
print ("stima grammi di solidi rimossi a 4.25 ore:", round(x_req, 2))
```
