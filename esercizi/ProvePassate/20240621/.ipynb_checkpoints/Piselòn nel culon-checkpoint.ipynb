```python
import numpy as np

def exp_rand(flamb, n):
    y = np.zeros(n)
    for i in range (n):
        X = np.random.rand()
        y[i] = -(np.log(1 - X) / flamb)

    return y
```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import probplot

lamb = 4
n = 10000

y_hist = exp_rand(lamb, n)

plt.hist(y_hist, density = True)

x_graph = np.linspace(min(y_hist), max(y_hist), n)
y_graph = expon.pdf(x_graph, 0, 1/lamb)
plt.plot(x_graph, y_graph, label = "Distribuzione teorica")

plt.title("Distribuzione numeri generati randomicamente con legge EXP(4)")
plt.legend()
plt.show()

print(" ")
probplot(y_hist, dist = expon, plot = plt)
plt.show()
```
