```python
import numpy as np
n = 12
x_values = np.array([32, 48, 72, 64, 48, 16, 40, 48, 48, 24, 80, 56])
y_values = np.array([230, 262, 323, 298, 255, 199, 248, 279, 267, 214, 359, 305])
```

```python
# Punto 1) b0 b1, [infb0, supb0] [infb1, supb1] con alpha = 0.05, coefficiente di determinazione R²
from scipy.stats import t

x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

sigma_xy = (1/n) * np.sum((x_values - x_mean) * (y_values - y_mean))
sigma_quadro_x = (1/n) * np.sum((x_values - x_mean)**2)

b1 = sigma_xy / sigma_quadro_x
b0 = y_mean - (b1 * x_mean)
print("b0:", round(b0, 5), ", b1:", round(b1, 5))

y_hat = b0 + (b1 * x_values)
r = y_values - y_hat
s_quadro = (1/(n-2)) * np.sum(r**2)
s = np.sqrt(s_quadro)

alpha = 0.05
T = t.ppf(1-alpha/2, n-2)

infb0 = b0 - (s * T * np.sqrt((1/n) + ((x_mean**2) / (n * sigma_quadro_x))))
supb0 = b0 + (s * T * np.sqrt((1/n) + ((x_mean**2) / (n * sigma_quadro_x))))
print("Intervallo di confidenza per b0: [", round(infb0,5), ",", round(supb0, 5), "]")

infb1 = b1 - (T * (s / np.sqrt(sigma_quadro_x * n)))
supb1 = b1 + (T * (s / np.sqrt(sigma_quadro_x * n)))
print("Intervallo di confidenza per b1: [", round(infb1,5), ",", round(supb1, 5), "]")

sigma_quadro_r = (1/n) * np.sum(r**2)
sigma_quadro_f = (sigma_xy**2) / sigma_quadro_x
sigma_quadro_y = sigma_quadro_f + sigma_quadro_r
R_quadro = 1 - (sigma_quadro_r / sigma_quadro_y)
print("Coefficiente di determinazione R²:", round(R_quadro, 5))
```

```python
# Punto 2) grafico retta + residui
import matplotlib.pyplot as plt

plt.title ("Grafico della durezza nel tempo della plastica")
plt.xlabel ("Tempo [h]")
plt.ylabel ("Durezza [Brinell]")

x_graph = np.linspace(min(x_values), max(x_values), 1000)
y_graph = b0 + (b1 * x_graph)

plt.plot(x_values, y_values, "*", label = "Valori originali")
plt.plot(x_graph, y_graph, label = "Retta di regressione")

plt.legend()
plt.show()
```

```python
# Punto 3) durezza a 48 ore
y = b0 + (b1 * 48)
print("La durezza stimata a 48 ore è", round(y, 5))
```

```python
# Punto 4) test indipendenza con alpha = 0.05

test = (b1 / s) * np.sqrt(n * sigma_quadro_x)

if np.abs(test) > T:
    print("L'ipotesi può essere rigettata")
else:
    print("L'ipotesi non può essere rigettata")
```
