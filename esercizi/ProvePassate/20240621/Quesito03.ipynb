```python
import numpy as np
lamb = 4

def exp_randgen(flamb, n):
    y = np.zeros(n)
    for i in range(n):
        X = np.random.rand()
        y[i] = -(np.log(1-X) / flamb)

    return y
```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import probplot

n = 10000
y_hist = exp_randgen(lamb, n)

plt.hist(y_hist, density = True)

x_graph = np.linspace(min(y_hist), max(y_hist), n)
y_graph = expon.pdf(x_graph, 0, 1/lamb)

plt.plot(x_graph, y_graph)

plt.title ("Generazione di numeri casuali secondo la distribuzione EXP(4)")
plt.xlabel ("Valori")
plt.ylabel ("Probabilità")

plt.show()

print (" ")
probplot(y_hist, dist = expon, plot = plt)
plt.show()
```
