```python
# Esempio di istogramma
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
edges = np.array([[60, 62],[63,65],[66,68],[69,71],[72,74]])
```

```python
ampiezza = edges[:,1] - edges[:,0]
```

```python
centers = 0.5*(edges[:,0] + edges[:,1])
```

```python
freq = np.array([5, 18, 42, 27, 8])
```

```python
# crea istogramma
fig, ax = plt.subplots(1, 1)
plt.bar(centers, freq, width=ampiezza)

# set titolo
ax.set_title("Indagine sul peso di 100 studenti")
  
# set assi
ax.set_xlabel('peso (kg)')
ax.set_ylabel('frequenza')

plt.show()
```
