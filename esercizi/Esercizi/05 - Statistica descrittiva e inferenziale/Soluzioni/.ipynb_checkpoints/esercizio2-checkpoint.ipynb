```python
#L'altezza di 2000 individui di una popolazione Ã¨ riportata nel file 'Data_altezze.dat'
```

```python
#1: calcola media e deviazione standard della popolazione
```

```python
import numpy as np

X = np.loadtxt('Data_altezze.dat') #array contente i dati del file --> le altezze
media = np.mean(X)
devStandard = np.std(X) 

print(
    "la media e la deviazione standard della popolazione sono rispettivamente: ",
    media,
    devStandard
)
```

```python
#2: costruire un istogramma a 20 barre
```

```python
import matplotlib.pyplot as plt

plt.hist(X,20, edgecolor="black")

plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.title("Indagine sull'altezza di una popolazione")

plt.show()
```

```python
#3: Ã¨ possibile adattare ai dati una distribuzione normale
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

plt.hist(X, 40, edgecolor="black", density="True")

miu = media
devSta = devStandard
x = np.arange(miu - (3*devSta) , miu + (3*devSta))
y = norm.pdf(x,miu,devSta)

plt.plot(x,y)

plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.title("Distribuzione normale")

plt.show()
```
