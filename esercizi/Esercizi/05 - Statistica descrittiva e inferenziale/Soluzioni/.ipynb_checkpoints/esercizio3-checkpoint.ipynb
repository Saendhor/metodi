```python
import matplotlib.pyplot as plt
import numpy as np

X = np.loadtxt("Data_altezze.dat")

plt.boxplot(X)

plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.title("Box Plot delle altezze di una popolazione")

plt.show()
```
