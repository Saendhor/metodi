```python
#un monitoraggio sulla densità di polveri sottili nell'aria ha condotto alle seguenti frequenze, in oppurtune unità di misura:
# classe [0,0.5[ | frequenza 10
# classe [0.5,1[ | frequenza 8
# classe [1,1.5[ | frequenza 6
# classe [1.5,2[ | frequenza 3

#Si calcolino i quartili emipirici.
```

```python
import matplotlib.pyplot as plt
import numpy as np

#asse x --> la classe
assex = np.array([[0,0.5], [0.5,1], [1,1.5], [1.5,2]])
#asse y --> le frequenze
assey = [10, 8, 6, 3]

#calcoliamo ampiezza e centri delle colonne
ampiezza = assex[:,1]-assex[:,0]
centro = (assex[:,1]+assex[:,0])/2 #fa la media delle due colonne

print ("Questi sono i valori dell'array centro")
print (centro)
plt.bar(centro, assey, ampiezza)
plt.xlabel("valore polveri")
plt.ylabel("frequenze")
plt.title("indagine densità polveri sottili")

plt.show()
```

```python
#funzione di ripartizione empirica 

#PASSO1 : CALCOLARE L'INDICE j 

import numpy as np

#asse x --> la classe
assex = np.array([[0,0.5], [0.5,1], [1,1.5], [1.5,2]])
#asse y --> le frequenze
assey = [10, 8, 6, 3]

#x 
x = [0, 0.5, 1, 1.5, 2] 

#valori di alpha -> primo, secondo e terzo quartile
alpha = [0.25, 0.50, 0.75]

#calcoliamo gli indici per la funzione di ripartizione empirica
indici = np.zeros(3, dtype = int)
print ("Il valore dell'array indici: ", indici)

#calcoliamo la probabilità che corrisponde a pi -> frequenza empirica
probabilità = assey / np.sum(assey)
assey = np.insert (assey, 0, 0)
assey = assey/np.sum (assey)
#calcoliamo la frequenza cumulativa -> somma del valore di pi con il valore della cumulativa trovato precedentemente
#nella formula sommaCumulativa corrisponde a F
sommaCumulativa = np.cumsum(probabilità)
print ("Valori in somma cumulativa (1)", sommaCumulativa)
sommaCumulativa = np.insert(sommaCumulativa,0,0) #inseriamo 0 nella prima posizione dell'array
print ("Valori in somma cumulativa (2)", sommaCumulativa)

import matplotlib.pyplot as plt
plt.title ("Distribuzione")
plt.plot (x, assey)
plt.plot (x, assey, "*")
plt.show ()


#scorro l'array alpha -> 3 elementi uso l'indice k
#scorro l'array contente le frequenze cumulative -> 5 elementi uso l'indice j
#se alpha è compreso fra Fj-1 e Fj allora inseriamo l'indice j nell'array degli indici 
for k in range(3):
    for j in range(1,5):
        if alpha[k] > sommaCumulativa[j-1] and alpha[k] <= sommaCumulativa[j]:
            indici[k]=j

print(
    "gli indici sono: ",
    indici)

#PASSI2e3: CALCOLO DEI QUANTILI

q_alpha = np.zeros(3) #array di tre elementi inizializzati a zero

for k in range(3):
    j = indici[k]
    q_alpha[k] = x[j-1]+(alpha[k]-sommaCumulativa[j-1])*((x[j]-x[j-1])/(sommaCumulativa[j]-sommaCumulativa[j-1]))

print(
    "i quantili sono: ",
    q_alpha)
    
```

```python
#creiamo il grafico

import matplotlib.pyplot as plt  

plt.plot(x, sommaCumulativa, "*", label="Frequenza relativa cumulativa")
plt.plot(x, sommaCumulativa, label="Funzione di ripartizione empirica")
plt.plot(q_alpha, alpha, "o", label="Quantili")

plt.legend()

plt.title("Indagine densità polveri sottili")
plt.xlabel("valore polveri")
plt.ylabel("Funzione di ripartizione empirica")

plt.show()
```
