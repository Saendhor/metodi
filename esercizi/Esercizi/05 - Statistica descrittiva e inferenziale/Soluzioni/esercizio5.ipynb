```python
#2: determinare l'intervallo di confidenza al 95%
```

```python
import numpy as np
from math import sqrt
from scipy.stats import t

X = np.array([19.8, 10.1, 14.9, 7.5, 15.4, 15.4, 15.4, 18.5, 7.9, 12.7, 11.9, 11.4, 11.4, 14.1, 17.6, 16.7, 15.8, 19.5, 8.8, 13.6, 11.9, 11.4])
n = X.size
alpha = 0.05
media = np.mean(X)
t = t.ppf(1-alpha/2, df=n-1)

S = np.std(X, ddof = 1)

min = media - S/sqrt(n) * t
max = media + S/sqrt(n) * t

print(
    "l'intervallo di confidenza al 95% è [{},{}]".format(min,max)
)
```
