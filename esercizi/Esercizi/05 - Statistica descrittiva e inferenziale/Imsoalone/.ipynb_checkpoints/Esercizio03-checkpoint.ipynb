```python
# L'altezza di 2000 individui di una popolazione Ã¨ riportata nel file "Data_altezze.dat"
import numpy as np
import matplotlib.pyplot as plt

#Carico i dati dal file .dat fornito
popolazione = np.loadtxt("Data_altezze.dat")
```

```python
# Costruiamo il box-plot

plt.boxplot(popolazione)
plt.title ("Indagine sull'altezza di una popolazione di 2000 individui")
plt.xlabel ("Altezze [cm]")
plt.ylabel ("Frequenze")

plt.show ()
```
