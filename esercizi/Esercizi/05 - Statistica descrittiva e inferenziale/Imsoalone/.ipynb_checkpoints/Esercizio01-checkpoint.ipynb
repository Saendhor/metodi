```python
#Un monitoraggio sulla densità di polveri sottili nell'aria ha condotto alle frequenze riportate di seguito.
# Si calcolino i quartili empirici.
```

```python
import numpy as np
import matplotlib.pyplot as plt

classi = np.array([[0, 0.5], [0.5, 1], [1, 1.5], [1.5, 2]])
frequenze = [10, 8, 6 ,3]

print ("Le classi da analizzare sono:", classi)
print ("Le frequenze date sono:", frequenze)
```

```python
# Passo 1: determiniamo l'indice j tale che F[j-1] < alpha <= F [j]

alpha = [0.25, 0.5, 0.75] # i quartili

asse_x = x = [0, 0.5, 1, 1.5, 2] #semplificazione dei valori delle classi
asse_y = frequenze
asse_y = np.insert(asse_y, 0, 0)

probabilità = asse_y / np.sum(asse_y)
print ("I valori di probabilità nei diversi punti dell'asse y sono ", probabilità)
print ("")

#grafico della distribuzione
plt.title ("Grafico della distribuzione di probabilità")
plt.plot (asse_x, probabilità)
plt.plot (asse_x, probabilità, "*")
plt.show()

F = np.cumsum (probabilità)
print ("")
print ("I valori della funzione di ripartizione empirica sono ", F)

indici = [0, 0, 0]
for k in range (3):
    for j in range (1,5):
        if alpha [k] > F [j-1] and alpha [k] <= F [j]:
            indici [k] = j #perchè voglio che l'indice di j (quindi di x) corrispondente al relativo quartile ricercato

print ("Gli indici ricavati sono", indici)
print ("")

# Passo 2 e 3: determiniamo i valori dei quantili di ordine alpha

q_alpha = [0, 0, 0]

for k in range (3):
    j = indici [k]
    q_alpha[k] = x[j-1] + (alpha[k] - F[j-1]) * (x[j] - x[j-1]) / (F[j] - F[j-1])

print ("I valori assunti da q_alpha sono", q_alpha)

plt.title ("Indagine polveri sottili")
plt.xlabel ("Valore polveri")
plt.ylabel ("Funzione di ripartizione empirica")
plt.plot(x, F, '*', label = "Frequenza relativa cumulativa")
plt.plot(x, F, label = "Funzione di ripartizione empirica")
plt.plot(q_alpha, alpha, "o", label = "Quantili di ordine alpha")

plt.legend()
plt.show()
```
