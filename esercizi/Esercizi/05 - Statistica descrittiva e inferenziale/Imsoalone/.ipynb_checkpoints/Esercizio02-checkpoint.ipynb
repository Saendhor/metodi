```python
# L'altezza di 2000 individui di una popolazione è riportata nel file Data_altezze.dat
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#Carico il file della popolazione e lo inserisco in un array apposito
popolazione = np.loadtxt("Data_altezze.dat")
```

```python
# Punto 1: Calcolare media e deviazione standard della popolazione

media = np.mean(popolazione)
devStandard = np.std(popolazione)

print ("La l'altezza media della popolazione è ", media, "mentre la deviazione standard è ", devStandard)
```

```python
# Punto 2: Costruire un istogramma a 20 barre

plt.title ("Indagine altezza popolazione di 2000 individui")
plt.xlabel ("Altezze [cm]")
plt.ylabel ("Frequenze")
plt.hist(popolazione, bins = 20, edgecolor = "black")
plt.show()
```

```python
# Punto 3: È possibile adattare ai dati una distribuzione normale?

plt.hist (popolazione, bins = 40, edgecolor = "black", density = "True")

x = np.arange (media - (3*devStandard), media + (3*devStandard))
y = norm.pdf (x, media, devStandard)
plt.plot (x,y) #distribuzione normale

plt.title ("Indagine altezza popolazione di 2000 individui")
plt.xlabel ("Altezze [cm]")
plt.ylabel ("Frequenze")

plt.show()
```
