```python
# Viene effettuato un tesr di rottura di un certo materiale ottenendo i seguenti valori
# in megapascal [MPa]

campioni = [19.8, 10.1, 14.9, 7.5, 15.4, 15.4, 15.4, 18.5, 7.9, 12.7, 11.9, 11.4, 11.4, 14.1, 17.6, 16.7, 15.8, 19.5, 8.8, 13.6, 11.9, 11.4]
n = campioni.size
# Dopo aver verificato graficamente che il campione proviene da una popolazione distribuita approssimativamente
# in modo normale, determinare l'intervallo di confidenza al 95% per la media.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import probplot
from scipy.stats import t
```

```python
# Parte grafica
plt.boxplot (campioni)
plt.show ()
print (" ")
probplot (campioni, dist = norm, plot = plt)
plt.show ()

#Esercizio
alpha = 0.05
media = np.mean (campioni)
t = t.ppf (1-alpha/2, df = n-1)

S = np.std (campioni)

lim_inf = media - (S / np.sqrt(n) * t)
lim_sup = media + (S / np.sqrt(n) * t)
```
