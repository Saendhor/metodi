```python
#Si suppone che l'altezza degli uomini in Italia segua approssimativamente una variabile aleatoria normale con 
# media 175cm e deviazione standard 9 cm.
```

```python
#1: Quale sarebbe la percentuale di italiani di statura superiore al metro e 90 (190cm)?
```

```python
#DISTRIBUZIONE NORMALE 
# la media è uguale a miu --> media = 175 = miu
# la deviazione standard è la radice quadrata della varianza --> devStandard = 9 = 9*9 = 81 = varianza
```

```python
from math import e
from math import sqrt
from scipy.stats import norm


x1 = 190 
miu1 = 175 
devStandard1 = 9
var1 = 81

#pencentuale di italiani sotto il metro e novanta --> P(X<190)
p1 = norm.cdf(x1,miu1,devStandard1)

#percentuale di italiani sopra il metro e novanta --> P(X>190) = 1-P(X<190)
p_es1 = 1-p1

print(
    "la percentuale di italiani di statura superiore al metro e 90 (190cm) è: ",
    round(p_es1,4)
)
                                    
```

```python
#2: Quale sarebbe la percentuale di reclute scartate alla visita di leva?
```

```python
from scipy.stats import norm

x2 = 153
miu2 = 175
devStandard2 = 9
var2 = 81

p_es2 = norm.cdf(x2,miu2,devStandard2)

print( 
    "la percentuale di reclute di altezza inferiore a 153cm scartate alla visita di leva è: ",
    round(p_es2,4)
)
```

```python
#Costruisci il grafico della funzione di densità 
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

miu = 175
devStandard = 9
x = np.arange(miu - (3*devStandard) , miu + (3*devStandard))
y = norm.pdf(x,miu, devStandard)
plt.plot(x,y)
plt.xlabel("altezza")
plt.ylabel("percentuale")
plt.title("Grafico della densità")
plt.show()
```
