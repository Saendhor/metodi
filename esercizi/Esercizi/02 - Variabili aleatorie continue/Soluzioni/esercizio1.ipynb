```python
#Il numero di anni di funzionamento di una radio ha una distribuzione esponenziale di parametro lambda = 1/8.
```

```python
from math import e
from scipy.stats import expon
```

```python
#1: Qual è la probabilità che una radio funzioni per più di dieci anni?
```

```python
#distribuzione Esponenziale --> occorre usare la cdf 
```

```python
x1 = 10 #valore per cui la probabilità deve essere maggiore 
lam1 = 1/8

#P(X<10) --> probabilità che la radio duri meno di 10 anni 
p1 = 1 - (pow(e,-lam1*x1)) #cdf 

#adesso da P(X>10) = 1-P(X<10) --> dall'evento certo togliamo il caso in cui la radio dura meno di 10 anni
p_es1 = 1 - p1

print(
    "la probabilità che una radio funzioni per più di dieci anni è: ",
    round(p_es1,4)
)
```

```python
#Costruire il grafico della densità
```

```python
import matplotlib.pyplot as plt
from math import e
from scipy.stats import expon

lam = 1/8
x = range(11)
y = expon.pdf(x,0,1/lam)
plt.plot(x,y)
plt.xlabel("tempo funzionamento")
plt.ylabel("probabilità")
plt.title("Grafico della densità")
plt.show()
```

```python
#Costruire il grafico della funzione di ripartizione della distribuzione
```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon

lam = 1/8
x = range(11)
y = expon.cdf(x,0,1/lam)
plt.plot(x,y)
plt.xlabel("tempo funzionamento")
plt.ylabel("probabilità")
plt.title("Grafico della funzione di ripartizione")
plt.show()
```
