```python
# Si suppone che l’altezza degli uomini in Italia segua approssimativamente una v.a. normale di media 175 cm e deviazione standard 9 cm.
from scipy.stats import norm
mi = 175 #cm
devStandard = 9 #cm
```

```python
#a) Quale sarebbe la percentuale di italiani di statura superiore al metro e 90?
x = 190
p = norm.cdf(x, mi, devStandard)
q = 1-p
print ("La percentuale di italiani di statura superiore al metro e 90 è", round(q, 2)*100, "%")
```

```python
#b) Alla visita di leva vengono scartate le reclute di altezza inferiore ai 153 cm. Quale sarebbe la percentuale di reclute scartate alla visita di leva?
x = 153
p2 = norm.cdf(x, mi, devStandard)
print ("La percentuale di reclute scartate alla visita di leva poichè di altezza inferiore ai 153 cm è", round(p2,2)*100, "%")
```

```python
#c) Costruire il grafico della densità e della funzione di ripartizione della distribuzione in esame.
import numpy as np
import matplotlib.pyplot as plt

x_graph = np.arange(100, 240)
y_graph1 = norm.pdf(x_graph, mi, devStandard)

plt.plot(x_graph, y_graph1)
plt.title("Distribuzione di probabilità")
plt.xlabel("Altezze [cm]")
plt.ylabel("Probabilità relativa")
plt.show()

print(" ")
y_graph2 = norm.cdf(x_graph, mi, devStandard)
plt.plot(x_graph, y_graph2)
plt.title("Funzione di ripartizione")
plt.xlabel("Altezze [cm]")
plt.ylabel("Probabilità relativa cumulativa")
plt.show()
```
