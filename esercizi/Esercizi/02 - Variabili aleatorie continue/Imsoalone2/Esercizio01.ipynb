```python
# Il numero di anni di funzionamento di una radio ha una distribuzione esponenziale di parametro λ=1/8.
from scipy.stats import expon
lamb = 1/8
```

```python
#a) Qual è la probabilità che una radio funzioni per più di dieci anni?
x = 10
p = expon.pdf(x, 0, 1/lamb)
print ("la probabilità che una radio funzioni per più di dieci anni è", round(p, 2))
```

```python
#b) Costruire il grafico della densità e della funzione di ripartizione della distribuzione in esame.
import numpy as np
import matplotlib.pyplot as plt

x_graph = np.arange(0, 20)
y_graph1 = expon.pdf(x_graph, 0, 1/lamb)
plt.plot(x_graph, y_graph1)
plt.title("Distribuzione di probabilità")
plt.xlabel("Tempo [Anni]")
plt.ylabel("Probabilità in funzione degli anni")
plt.show()

print(" ")
y_graph2 = expon.cdf(x_graph, 0, 1/lamb)
plt.plot(x_graph, y_graph2)
plt.title("Funzione di ripartizione di EXP(lamb)")
plt.xlabel("Tempo [Anni]")
plt.ylabel("Probabilità cumulativa")
plt.show()
```
