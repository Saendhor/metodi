```python
# Il numero di anni di funzionamento di una radio ha una distribuzione esponenziale di parametro
# λ=1/8
```

```python
# Punto a) Qual è la probabilità che una radio funzioni per più di dieci anni?
from scipy.stats import expon
from math import e

x = 10
lamb = 1/8

result_a = 1 - (1 - pow(e, -lamb * x))
print ("La probabilità che una radio funzioni per più di dieci anni è ", round(result_a, 3))
```

```python
# Punto b) Costruire il grafico della densità e della funzione di ripartizione
# della distribuzione in esame

import matplotlib.pyplot as plt
import numpy as np

x = range (11) #definisco un range arbitrario

#Densità
y = expon.pdf(x, 0, 1/lamb)
plt.title ("Grafico della densità")
plt.plot (x, y)
plt.show ()

#Ripartizione
y = expon.cdf(x, 0, 1/lamb)
plt.title ( "Grafico della funzione di ripartizione")
plt.plot (x, y)
plt.show ()
```
