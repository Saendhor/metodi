```python
# Si suppone che l’altezza degli uomini in Italia segua approssimativamente una v.a.
# normale di media 175 cm e deviazione standard 9 cm.

mi = 175 #µ
sig = 9 #sigma²
```

```python
# Quale sarebbe la percentuale di italiani di statura superiore al metro e 90? 
from scipy.stats import norm

x = 190

result_a =  (1 - norm.cdf (x, mi, sig))
print ("La percentuale di italiani di statura superiore al metro e 90 è del", round(result_a, 3)*100, "%" )
```

```python
# Alla visita di leva vengono scartate le reclute di altezza inferiore ai 153 cm.
# Quale sarebbe la percentuale di reclute scartate alla visita di leva? 

x = 153

result_b  = norm.cdf (x, mi, sig)
print ("La percentuale di reclute scartate alla visita di leva è del", round(result_b, 3)*100, "%" )
```

```python
# Costruire il grafico della densità e della funzione di ripartizione della
# distribuzione in esame.
import matplotlib.pyplot as plt
import numpy as np

x = np.arange (mi - (3*sig), mi + (3*sig))

# Problema 1
y = norm.pdf (x, mi, sig)

plt.title ("Grafico densità")
plt.plot (x, y)
plt.show ()

# Problema 2
y = norm.cdf (x, mi, sig)

plt.title ("Grafico funzione di ripartizione")
plt.plot (x, y)
plt.show ()
```
