```python
# La probabilità di contrarre una malattia rara è dello 0.03%. Qual è la probabilità che in una città dove
# vivono 20000 persone vi siano meno di 4 persone che contraggono la malattia? Costruire il grafico
# della densità e della funzione di ripartizione della distribuzione in esame.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

p = 0.03 / 100
n = 20000
k = 3 +1 # meno di 4 persone

y = 0
for i in range(k):
    y = y + binom.pmf(i, n, p)
print ("la probabilità che in una città dove vivono 20000 persone vi siano meno di 4 persone",
       "che contraggono la malattia è", y)

x_graph = np.arange(0, 30)
y_graph = binom.pmf(x_graph, n, p)
plt.plot (x_graph, y_graph)
plt.title ("Distribuzione di probabilità")
plt.xlabel("Campioni")
plt.ylabel("Probabilità per numero di campioni")
plt.show()

print (" ")
y2_graph = binom.cdf(x_graph, n, p)
plt.plot (x_graph, y2_graph)
plt.title ("Funzione di ripartizione")
plt.xlabel("Campioni")
plt.ylabel("Probabilità cumulativa per numero di campioni")
plt.show()
```
