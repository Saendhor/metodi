```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
# "meno di 4 persone"
```

```python
n = 20000
```

```python
p = 0.0003
```

```python
result = 1 - ( binom.pmf (3, n, p) + binom.pmf (2, n, p) + binom.pmf (1, n, p) + binom.pmf (0, n, p))
```

```python
display (result)
```

```python
x = np.arange(0, 25)
```

```python
y = binom.pmf (x, n, p)
```

```python
#plt.plot (x, y) # è discreta ma tracciamo comunque la linea
plt.plot (x, y, '*')
plt.show ()
```

```python
z = np.cumsum (y) # per le distribuzioni è la sommatoria!
```

```python
plt.plot (x, z)
plt.plot (x, z, '*')
plt.plot (3, 1-result, '.')
plt.show ()
```
