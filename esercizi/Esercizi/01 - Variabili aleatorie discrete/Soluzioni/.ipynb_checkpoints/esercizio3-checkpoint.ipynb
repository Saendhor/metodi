```python
from math import e
from math import factorial
```

```python
#Qual è la probabilità che in una citta dove vivono 20000 persone vi siano meno di 4 persone che contraggono la malattia?
```

```python
#si deve usare la distribuzione di Poisson
```

```python
n = 20000 #numero di prove --> numero di persone
p = 0.03/100 #probabilità
lam1 = n*p
#il numero di successi k varia da un minimo di 0 a un massimo di 3
k0 = 0
k1 = 1
k2 = 2
k3 = 3

p0 = pow(e,-lam1) *((pow(lam1,k0))/(factorial(k0)))
p1 = pow(e,-lam1) *((pow(lam1,k1))/(factorial(k1)))
p2 = pow(e,-lam1) *((pow(lam1,k2))/(factorial(k2)))
p3 = pow(e,-lam1) *((pow(lam1,k3))/(factorial(k3)))

p_es1 = p0 + p1 + p2 + p3


print(
    "la probabilità che in una città di 20000 persone vi siano meno di 4 persone che contraggono la malattia è: ",
    round(p_es1,4)
)

```

```python
#Costruisci il grafico della densità
```

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

x = range(0, n+1)
n = 20000
p = 0.03/100 
lamD = n*p
y = poisson.pmf(x,lamD)
plt.plot(x,y,".")
plt.xlabel("persone")
plt.title("Grafico della densità")
plt.ylabel("probabilità")
plt.show()
```
