```python
import matplotlib.pyplot as plt
import numpy as np

x = np.loadtxt("Data_altezze.dat")

media = np.mean(x)
dev_st = np.std(x)

print("media : ",media)
print("dev_std : ",dev_st)
```

```python
plt.hist(x,20,edgecolor = "black")

plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.title("indagini sull'altezza")

plt.show()
```

```python
plt.hist(x,40,edgecolor="black",density="True")

from scipy.stats import norm

x = np.arange(media-(3*dev_st),media+(3*dev_st))
y = norm.pdf(x,media,dev_st)
print(y)
plt.plot(x,y)
plt.xlabel("altezze")
plt.ylabel("freq")
plt.title("indagine altezza")

plt.show()
```
