```python
import numpy as np

def fun (x):
    return (1+np.cos(x))/(2*np.pi)

def rigetto(a,b,M):#M rappresenta il massimo valore assunto dalla funzione
    while True:
        r1 = np.random.rand()
        r2 = np.random.rand()
        xi = a + r1*(b-a)
        eta = M*r2
        if eta<=fun(xi):
            break
    return xi
    


a = -np.pi
b = np.pi
M = 1/np.pi

N = 100000
x = np.zeros(N)

for i in range(N):
    x[i] = rigetto(a,b,M)
```

```python
import matplotlib.pyplot as plt

plt.hist(x,100, density = True)
xx = np.linspace(a,b,1000)
yy = fun(xx)
plt.plot(xx,yy)
plt.show()
```
