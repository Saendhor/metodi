```python
#esponenziale

n = 1000
lam = 3

x = np.random.rand(n)

e = -np.log(x)/lam

xx = np.linspace(0,10,1000)
from scipy.stats import expon
yy = expon.pdf(xx,0,1/lam)
plt.plot(xx,yy)

plt.hist(e,100,density=True)
plt.show()
```

```python
#normale

var = 2
media = 1

x = np.random.rand(n)

x1 = x[0:int(n/2)]
x2 = x[int(n/2):n]

eta1 = np.sqrt(-2*np.log(x1))*np.cos(2*np.pi*x2)
eta2 = np.sqrt(-2*np.log(x1))*np.sin(2*np.pi*x2)

n_ = np.zeros(n)
n_ [0:int(n/2)]= eta1
n_[int(n/2):n] = eta2

N_ = media + np.sqrt(var)*n_

plt.hist(N_,20,density=True,edgecolor="black")

xx = np.linspace(-6,7,1000)
from scipy.stats import norm
yy = norm.pdf(xx,media,np.sqrt(var))
plt.plot(xx,yy)
plt.show()
```

```python
probplot(N_,dist=norm,plot=plt)
plt.show()
```
