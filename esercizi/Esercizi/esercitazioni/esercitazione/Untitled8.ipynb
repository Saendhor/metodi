```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

lam = 2
n0 = 2
mu = 1
T = 5

t = np.array([0])
x = np.array([n0])
i = 0
time = 0

while time <= T:
    if x[i]==0:
        break
    r = np.random.rand()
    time = time - np.log(r)/(x[i]*lam+x[i]*mu)
    s = np.random.rand()
    if s < lam/(lam+mu):
        x = np.append(x,x[i]+1)
    else:
        x = np.append(x,x[i]-1)
    i = i+1
    t = np.append(t,time)

plt.plot(t,x)
plt.show()
```
