```python
#esercizio 2
import matplotlib.pyplot as plt
n = 10
x = np.array([2.5,3,3.5,4,4.5,5,5.5,6,6.5,7])
y = np.array([4.3,1.5,1.8,4.9,4.2,4.8,5.8,6.2,7,7.9])
x_ = np.mean(x)
y_ = np.mean(y)

sig_xy = np.sum((x-x_)*(y-y_))/n
sig_2x = np.sum((x-x_)**2)/n

b0 = y_ - sig_xy/sig_2x*x_
b1 = sig_xy/sig_2x

yy = b0 + x*b1


plt.plot(x,y,'*')
plt.plot(x,yy)
plt.show()
```

```python
y_hat = b0 + b1*x
r = y - y_hat

s2 = np.sum(r**2)/(n-2)

sig_r2 = np.sum(r**2)/n
sig_f2 = sig_xy**2/sig_2x

sig_y2 = sig_f2 + sig_r2

R2 = 1 - sig_r2/sig_y2
display(R2)
```

```python
alpha = 0.05
from scipy.stats import t
s = np.sqrt(s2)
phi = t.ppf(1-alpha/2,n-2)
il = b0 - s*np.sqrt(1/n+x_**2/n*sig_2x)*phi
ir = b0 + s*np.sqrt(1/n+x_**2/n*sig_2x)*phi
display(il,ir)

il_ = b1 - s/(np.sqrt(sig_2x)*np.sqrt(n))*phi
ir_ = b1 + s/(np.sqrt(sig_2x)*np.sqrt(n))*phi
display(il_,ir_)
```

```python
#esercizio 3

def fun(x):
    return -3/4*x**2+3/2*x
    

def rigetto(a,b,m):
    while True:
        r = np.random.rand()
        s = np.random.rand()
        xi = a +r*(b-a)
        eta = s*m
        if eta < fun(xi):
           break
    return xi

m = 0.75
a = 0
b = 2

N = 1000000

x = np.zeros(N)
for i in range(N):
    x[i] = rigetto(a,b,m)


display(np.max(x))

plt.hist(x,100,density=True,edgecolor="black")
xx = np.linspace(0,2,100)
yy = fun(xx)
plt.plot(xx,yy)
plt.show()
```
