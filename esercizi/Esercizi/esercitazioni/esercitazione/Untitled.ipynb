```python
import numpy as np
import matplotlib.pyplot as plt

x = np.loadtxt("Data_altezze.dat")
med = np.mean(x)
dev = np.std(x,ddof=1)
print("media: ",med)
print("deviazione standard: ",dev)
```

```python
plt.hist(x,20,edgecolor="black")
plt.title("statistiche altezze")
plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.show()
```

```python
from scipy.stats import skew
from scipy.stats import kurtosis

skew_ = skew(x)
print("skew: ",skew_)
kurt = kurtosis(x,fisher=False)
print("kurtosis: ",kurt)
```

```python
#dagli indici precedenti possiamo adattare una distribuzione normale
from scipy.stats import norm
xx = np.linspace(130,210,200)
yy = norm.pdf(xx,med,dev)
plt.hist(x,40,edgecolor="black",density = True)
plt.plot(xx,yy)
plt.title("statistiche altezze")
plt.xlabel("altezze")
plt.ylabel("frequenze")
plt.show()
```
