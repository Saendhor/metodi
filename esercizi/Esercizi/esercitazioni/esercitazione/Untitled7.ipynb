```python
import numpy as np
import matplotlib.pyplot as plt

n0 = 2
lam = 2
mu = 1
T = 5
```

```python
t = np.array([0.])
x = np.array([n0])
time = 0
i = 0

while time <= T:
    if x[i]==0:
        break
    r = np.random.rand()
    time = time - np.log(r)/(lam*x[i]+mu*x[i])
    s = np.random.rand()
    if s <= lam/(lam+mu):
        x = np.append(x,x[i]+1)
    else:
        x = np.append(x,x[i]-1)
    t = np.append(t,time)
    i = i+1

```

```python
plt.plot(t,x)
plt.show()
```
