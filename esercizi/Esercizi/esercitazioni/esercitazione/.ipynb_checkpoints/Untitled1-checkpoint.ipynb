```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0,0.5],[0.5,1],[1,1.5],[1.5,2]])
freq = np.array([10,8,6,3])

centri = (x[:,1]+x[:,0])/2
ampiezze = x[:,1]-x[:,0]

plt.bar(centri,freq,ampiezze)
plt.xlabel("polveri")
plt.ylabel("frequenza")
plt.title("indagine sulle polveri")
plt.show()
```

```python
#calcolare i quartili empirici

alpha = np.array([0.25,0.5,0.75])
n = np.sum(freq)
f = freq
```

```python
f = f/n
xx = np.cumsum(f)
xx = np.insert(xx,0,0)
```

```python
index = np.zeros(3,dtype=int)
for k in range(3):
    for j in range(1,5):
       if xx[j-1]<alpha[k] and xx[j]>=alpha[k]:
           index[k] = j

print(index)
```

```python
q = np.zeros(3)
vv = np.array([0,0.5,1,1.5,2])
for k in range(3):
    j = index[k]
    q[k] = vv[j-1] + (alpha[k] - xx[j-1]) * ( vv[j] - vv[j-1] ) / (xx[j] - xx[j-1])

print(q)
```

```python
plt.plot(vv,xx,'*')
plt.plot(vv,xx)
plt.plot(q,alpha,'o')
plt.show()
```
