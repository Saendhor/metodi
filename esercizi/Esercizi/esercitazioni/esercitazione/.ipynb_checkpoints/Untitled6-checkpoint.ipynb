```python
import numpy as np

DATA = np.loadtxt("DATA_reg_lin_2.dat")

x = DATA[:,0]
y = DATA[:,1]
n = x.size
e = np.ones(n)

X = np.zeros((n,2))
X[:,0]=e
X[:,1]=x

p = np.linalg.pinv(X)

b = np.dot(p,y)

xx = np.linspace(0,12,100)
yy = b[0] + b[1]*xx

import matplotlib.pyplot as plt
plt.plot(x,y,'*')
plt.plot(xx,yy)
plt.show()
```

```python
y_hat = np.dot(X,b)

r = y-y_hat
plt.plot(r,x,'*')
plt.show()
```

```python
s2 = np.sum(r**2)/(n-2)
s = np.sqrt(s2)

M = np.linalg.inv(np.dot(X.T,X))
m = M[1][1]

display(m)

```

```python
#test

tn = b[1]/(s*np.sqrt(m))
from scipy.stats import t
T = t.ppf(1-0.05/2,n-2)
display(tn,T)
```
