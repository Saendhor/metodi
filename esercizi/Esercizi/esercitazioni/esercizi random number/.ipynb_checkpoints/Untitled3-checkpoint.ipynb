```python
import numpy as np
from scipy.stats import expon

lam = 2
N = 20000
x = np.random.rand(N)

Y = -np.log(x)/lam

import matplotlib.pyplot as plt

xx = np.linspace(0,5,1000)
yy = expon.pdf(xx,0,1/2)

plt.hist(Y,100,density = "True")
plt.plot(xx,yy)
plt.show()
```

```python
from scipy.stats import probplot
probplot(Y,dist=expon,plot = plt)
plt.show()
```
