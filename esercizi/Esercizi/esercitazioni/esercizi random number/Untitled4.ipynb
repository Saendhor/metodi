```python
import numpy as np

N = 20000
x = np.random.rand(20000)

xi1 = x[0:int(N/2)]
xi2 = x[int(N/2):N]
```

```python
eta1 = np.sqrt(-2*np.log(xi1))*np.cos(2*np.pi*xi2)
eta2 = np.sqrt(-2*np.log(xi1))*np.sin(2*np.pi*xi2)
display(eta1,eta2)
```

```python
Y = np.zeros(N)
Y[0:int(N/2)] = eta1
Y[int(N/2):N] = eta2
```

```python
import matplotlib.pyplot as plt

xx = np.linspace(-5,5,1000)

from scipy.stats import norm

yy = norm.pdf(xx)

plt.hist(Y,100, density = True)
plt.plot(xx,yy)
plt.show()
```

```python
from scipy.stats import probplot
probplot(Y,dist=norm,plot=plt)
plt.show()
```
