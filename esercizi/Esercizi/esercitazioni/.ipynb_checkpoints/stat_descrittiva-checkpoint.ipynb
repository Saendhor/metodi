```python
import matplotlib.pyplot as plt
import numpy as np

ascissa = np.array([[0,0.5],[0.5,1],[1,1.5],[1.5,2]])
ordinata = [10,8,6,3]

ampiezza = ascissa[:,1] - ascissa[:,0]
#calcolo ampiezze colonne
display(ampiezza)
centro = (ascissa[:,1]+ascissa[:,0])/2
#calcolo centro colonne per istogramma
display(centro)

plt.bar(centro,ordinata,ampiezza)
plt.xlabel("valore polveri")
plt.ylabel("freq")
plt.title("indagine densità polveri sottili")

plt.show()
```

```python
#punto 2

x = [0,0.5,1,1.5,2]

alpha = [0.25,0.5,0.75]

indici = np.zeros(3,dtype=int)
display(indici)
```

```python
probabilità = ordinata/np.sum(ordinata)
display(probabilità)
```

```python
probC = np.cumsum(probabilità)
display(probC)
```

```python
probC = np.insert(probC,0,0)
```

```python
display(probC)
```

```python
for k in range(3):
    for j in range(1,5):
        if(alpha[k]>probC[j-1] and alpha[k]<=probC[j]):
            indici[k] = j

print(indici)
```

```python
q_alpha = np.zeros(3)

for k in range(3):
    j = indici[k]
    q_alpha[k] = x[j-1]+(alpha[k]-probC[j-1])*(x[j]-x[j-1])/(probC[j]-probC[j-1])

#in q_alpha ottengo i quantili empirici
print(q_alpha)
```

```python
plt.plot(x,probC,"*",label="freq relativa cumulativa")
plt.plot(x,probC,label="funzione di ripartizione empirica")
plt.plot(q_alpha,alpha,"o",label="quantili")

plt.legend()

plt.title("indagine densità polveri sottili")
plt.xlabel("valore polveri")
plt.ylabel("funz di ripartizione empirica")

plt.show()
```
