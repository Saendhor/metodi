```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([19.8,10.1,14.9,7.5,15.4,15.4,15.4,18.5,7.9,12.7,11.9,11.4,11.4,14.1,17.6,16.7,15.8,19.5,8.8,13.6,11.9,11.4])

plt.boxplot(x)
plt.show()
```

```python
from scipy.stats import probplot
from scipy.stats import norm

probplot(x,dist=norm,plot=plt)
plt.show()
```

```python
media = np.mean(x)
alpha = 0.05
from scipy.stats import t
from math import sqrt

S = np.std(x,ddof=1)
n = x.size
phi = t.ppf(1-alpha/2,n-1)
min = media - S/sqrt(n) * phi
max = media + S/sqrt(n) * phi

print("intervallo di confidenza al 95%: [{},{}],".format(min,max))
```
