```python
import matplotlib.pyplot as p
import numpy as np
from scipy.stats import norm
x = np.loadtxt("Data_altezze.dat")

p.hist(x,20,edgecolor="black")
p.show()

media = np.mean(x)
dev = np.std(x)
print(media)
print(dev)
```

```python
p.hist(x,40,edgecolor="black",density="True")
z = np.arange(media - (3*dev) , media + (3*dev))
y = norm.pdf(z,media,dev)
print(y)
p.xlabel("altezze")
p.ylabel("freq")
p.title("indagine altezza")

p.plot(z,y)
p.show()
```
