```python
import numpy as np
n0 = 2
lam = 2
mu = 1
T = 5

t = np.array([0.])
X = np.array([n0])
time = 0
i = 0

while time <= T:
    if X[i] == 0:
        break
    r = np.random.rand()
    time = time-np.log(r)/(lam*X[i]+mu*X[i])
    s = np.random.rand()
    if s < lam/(lam+mu):
        X = np.append(X,X[i]+1)
    else:
        X = np.append(X,X[i]-1)
    t = np.append(t,time)
    i = i+1

import matplotlib.pyplot as plt
plt.plot(t,X)
plt.show()
```
