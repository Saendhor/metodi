```python
#si scrivano delle funzione Python per generare numeri random con le seguenti distribuzioni:
#U[-1,1], EXP(3), N(1,2), X^2(5).
#Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche
#mediante istogramma e mediante gafico quantile-quantile(Q-Qplot)
```

```python
#1: distribuzione UNIFORME U[-1,1]
```

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot

N = 1000
#estremi dell'intervallo
a = -1
b = 1
#variabile X --> generiamo numeri pseudo-casuali con distribuzione uniforme
X = np.random.rand(N)
#calcolaimo la variabile Y --> Y = a + X*(b-a)
Y = a+X*(b-a)
#print("Y: ", Y)

#facciamo il grafico: istogramma
x = np.linspace(a,b,N)
y = uniform.pdf(x, a, b-a)
plt.hist(Y, density=True, edgecolor="black")
plt.plot(x,y)
plt.title("Distribuzione uniforme")
plt.show()


```

```python
#grafico quantile-quantile
probplot(Y, dist=uniform, plot=plt)
plt.show()
```

```python
#2: distribuzione ESPONENZIALE EXP(3)
```

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt

lamb = 3
N = 1000
#calcoliamo X
X = np.random.rand(N)
#calcolaimo Y --> Y = -logX/lambda con lambda > 0
Y = -np.log(X)/lamb

#grafico istogramma
x = np.linspace(0,10,N)
y = expon.pdf(x,0,1/lamb)
plt.title("Distribuzione esponenziale")
plt.plot(x,y)
plt.hist(Y, bins=100, density=True)
plt.show()
```

```python
#grafico quantile-quantile

from scipy.stats import probplot
probplot(Y, dist=expon, plot=plt)
plt.show
```

```python
#3: distribuzione NORMALE N(1,2)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
 
miu = 1 #media
var = 2 #sigma^2
N = 1000
#calcoliamo X
X = np.random.rand(N)
#calcoliamo e1 e e2 --> dividiamo in due l'arrayX
e1 = X[0: int(N/2)] #da indice 0 a N/2 (escluso)
e2 = X[int(N/2):N] #da N/2 a N

#calcoliamo eta1 e eta2
eta1 = np.sqrt(-2*np.log(e1))*np.cos(2*np.pi*e2)
eta2 = np.sqrt(-2*np.log(e1))*np.sin(2*np.pi*e2)

#calcoliamo Y
Y = np.zeros(N)
Y[0: int(N/2)] = miu + var *eta1
Y[int(N/2):N] = miu + var*eta2

#grafico istogramma
x = np.linspace(-6, 7, N)
y = norm.pdf(x,miu,var)
plt.plot(x,y)
plt.title("Distribuzione normale")
plt.hist(Y, bins=20, density= True, edgecolor="black")
plt.show()



```

```python
#grafico quantile-quantile
from scipy.stats import probplot
probplot(Y, dist=norm, plot=plt)
plt.show()
```
