```python
#generare numeri pseudo-casuali con distribuzione:
# f(x) = 1+cos(x)/2*pigreco con x appartenente a [-pigreco, +pigreco].
#Eseguire un confronto statistico tra i numeri generati e la distribuzione teorica.
```

```python
import numpy as np
import matplotlib.pyplot as plt 

#definiamo y che è la distribuzione data
def fun(x):
    y = (1 + np.cos(x))/(2*np.pi)
    return y

#definiamo la funzione rigetto
def rigetto(a,b,M):
    while True: #ciclo infinito fin quando non si trova un campione valido
        r1=np.random.rand() #numero casuale tra 0 e 1
        r2=np.random.rand() #numero casuale tra 0 e 1
        xi = a+r1*(b-a) #campione uniforme nell'intervallo [a,b[
        eta = M*r2 #campione scalato uniformemente
        if eta <= fun(xi):  #condizione per accetare il campione -> se lo trova esce dal ciclo e ritorna il campione
            break    
    return xi

#definiamo i parametri --> x appartiene a [-pigreco, +pigreco] che sono rispettivamente a e b dell'intervallo
a = -np.pi
b = np.pi
M = 1/np.pi 

N = 100000
X = np.zeros(N)
for i in range(N):
    X[i]=rigetto(a,b,M)


#implementiamo il grafico per il confronto statistico
xx = np.linspace(a,b,1000)
yy = fun(xx)
fig, ax = plt.subplots(1,1)
ax.hist(X, bins=100, density=True)

plt.plot(xx,yy)
plt.title("Distribuzione di f(x)")

plt.show()
```
