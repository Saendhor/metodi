```python
# Si scrivano delle funzioni di Python per generare numeri random con distribuzione
# Ï‡2(5).

# Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche
# mediante istogramma e mediante grafico quantile-quantile
n = 1000
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
from scipy.stats import probplot

arg = 5
m = n * arg

xi_quadro = np.zeros(m)

for i in range (int(m/2)):
    xi1 = np.random.rand()
    xi2 = np.random.rand()
    
    xi_quadro [i] = (np.sqrt(-2 * np.log(xi1)) * np.cos(2* np.pi * xi2))**2
    xi_quadro [i + int(m/2)] = (np.sqrt(-2 * np.log(xi1)) * np.sin(2* np.pi * xi2))**2

eta = np.zeros(n)

for i in range(n):
    eta [i] = xi_quadro[i*5] + xi_quadro[(i*5) +1] + xi_quadro[(i*5) +2] + xi_quadro[(i*5) +3] + xi_quadro[(i*5) +4]

plt.hist(eta, density = True)
x_graph = np.linspace(min(eta), max(eta), n)
y_graph = chi2.pdf(x_graph, arg)
plt.plot(x_graph, y_graph)

plt.title("Generazione di numeri casuali con distribuzione Ï‡2(5)")
plt.xlabel("Range di valori generati")
plt.ylabel("Distribuzione di probabilitÃ ")

plt.show()

print(" ")
probplot(eta, dist = chi2(arg), plot = plt)
plt.show()
```
