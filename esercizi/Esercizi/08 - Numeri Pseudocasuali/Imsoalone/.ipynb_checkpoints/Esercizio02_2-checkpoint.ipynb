```python
# Si scriva una funzione di python che generi numeri random con distribuzione
# EXP(3)
lamb = 3
# Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche mediante
# istogramma e mediante frafico quantile-quantile
n = 1000
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import probplot

y = np.zeros(n)

for i in range (n):
    xi = np.random.rand()
    y [i] = (-1) * (np.log(1 - xi) / lamb)

plt.hist(y, density = True)
plt.title ("Distribuzione EXP(3)")
plt.xlabel ("Range valori") 
plt.ylabel ("Probabilità") 
x_graph = np.linspace (min(y), max(y), n)
y_graph = expon.pdf(x_graph, 0, 1/lamb)
plt.plot (x_graph, y_graph)
plt.show()

print(" ")
probplot(y, dist = expon, plot = plt)
plt.show()
```
