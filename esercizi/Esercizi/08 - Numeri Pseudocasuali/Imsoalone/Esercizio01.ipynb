```python
# Scrivere un algoritmo per generare numeri pseudo casuali con distribuzione multinomiale.
p = [1/4, 1/2, 1/4]

# Generare 2000 di questi numeri e costruire un istogramma verificando l'accordo con la distribuzione
# teorica di probabilità
n = 2000

import numpy as np
import matplotlib.pyplot as plt
```

```python
def rand_multi (p):
    F = np.cumsum (p)
    rand_num = np.random.rand()

    if rand_num < F[0]:
        return 1
    if rand_num < F[1]:
        return 2
    if rand_num < F[2]:
        return 3    
```

```python
X = np.zeros(n)
for i in range(n):
    X[i] = rand_multi (p)

prob_dim = len(p)
occorrenze = np.zeros(prob_dim)
for i in range (prob_dim):
    occorrenze [i] = np.sum(X == i+1) # se l'item i di X è == 1 allora somma in occorrenze [0] e così fino a X == 3

import matplotlib.pyplot as plt
x_graph = [1, 2, 3]
y_graph = occorrenze / n # probabilità empiriche

plt.bar(x_graph, y_graph)
plt.title ("Istogramma")
plt.xlabel ("Numero occorrenze")
plt.ylabel ("Probabilità")
plt.show ()
```
