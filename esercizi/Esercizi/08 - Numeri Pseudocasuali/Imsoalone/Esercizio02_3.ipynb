```python
#Si scrivano delle funzioni di Python per generare numeri random con distribuzione
# N (1, 2)
mi = 1
devStandard = 2
# Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche
# mediante istogramma e mediante grafico quantile-quantile
n = 1000
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import probplot

xi = np.zeros(n)

for i in range (int(n/2)):
    xi1 = np.random.rand()
    xi2 = np.random.rand()
    
    xi [i] = np.sqrt((-2) * np.log(xi1)) * np.cos(2* np.pi * xi2)
    xi [i + int(n/2)] = np.sqrt((-2) * np.log(xi1)) * np.sin(2* np.pi * xi2)

eta = np.zeros(n)

for i in range (n):
    eta [i] = mi + (np.sqrt(devStandard) * xi[i])

plt.hist(eta, density = True)
x_graph = np.linspace(min(eta), max(eta), n)
y_graph = norm.pdf(x_graph, mi, devStandard)
plt.title ("Numeri randomici generati secondo N(1, 2)")
plt.xlabel("Range di valori")
plt.ylabel("Distribuzione di probabilit")
plt.plot(x_graph, y_graph)
plt.show()

print (" ")
probplot(eta, dist = norm, plot = plt)
plt.show()
```
