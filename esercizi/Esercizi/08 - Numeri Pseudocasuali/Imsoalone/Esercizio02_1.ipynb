```python
# Si scriva una funzione di python che generi numeri random con distribuzione
# U ([-1, 1])
a = -1
b = 1
# Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche mediante
# istogramma e mediante frafico quantile-quantile
n = 1000
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot
from scipy.stats import uniform


y = np.zeros(n)

for i in range (n):
    rand = np.random.rand()
    y [i] = rand * (b-a) + a

plt.hist(y, density = True)

x_graph = np.linspace (-1, 1, n)
y_graph = uniform.pdf (x_graph, a, b-a,)

plt.xlabel ("Range valori assunti")
plt.ylabel ("Probabilit")
plt.plot(x_graph, y_graph)
plt.show()

print(" ")
probplot (y, dist = uniform, plot = plt)
plt.show()
```
