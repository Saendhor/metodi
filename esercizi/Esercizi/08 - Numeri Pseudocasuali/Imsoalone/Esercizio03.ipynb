```python
# Generare numeri pseudo-casuali con distribuzione

# f(x) = (1 + cos(x)) / 2pi con x â‚¬ [-pi, pi]
def f (x):
    return (1 + np.cos(x)) / (2* np.pi)
    
# Eseguire un confronto statistico tra i numeri generati e la distribuzione teorica.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rigetto (a, b, M):
    while True:
        r1 = np.random.rand()
        r2 = np.random.rand()
        xi = a + r1*(b-a)
        eta = M * r2
        
        if eta <= f(xi):
            break

    return xi
        
a = -np.pi
b = np.pi
M = (1 + np.cos(0)) / 2*np.pi

n = 1000
x = np.zeros(n)

for i in range(n):
    x[i] = rigetto(a, b, M)

plt.hist(x, density = True)
x_graph = np.linspace(a, b, n)
y_graph = f(x_graph)
plt.plot (x_graph, y_graph)
plt.show()
```
