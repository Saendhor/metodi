```python
#Si scrivano delle funzioni di Python per generare numeri random con distribuzione
#U ([−1, 1]), EXP(3), N (1, 2), χ2(5).

# Si generino N = 1000 di questi numeri e si confrontino i risultati con le distribuzioni teoriche
# mediante istogramma e mediante grafico quantile-quantile
n = 1000
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
from scipy.stats import expon
from scipy.stats import norm
from scipy.stats import probplot
from scipy.stats import uniform
```

```python
# Punto 1) U ([−1, 1])
a = -1
b = 1
y1 = np.zeros(n)

# Y = a + X(b-a)
for i in range(n):
    xi = np.random.rand()
    y1[i] = (2*(xi)) -1

# Istogramma
x1_graph = np.linspace( (-1), 1, n)
y1_graph = uniform.pdf(x1_graph, a, b-a)
plt.hist(y1, density = True)
plt.title ("Distribuzione uniforme in [a, b]")
plt.plot (x1_graph, y1_graph)
plt.show()

print(" ")
# Grafico quantile quantile
probplot(y1, dist = uniform, plot = plt)
plt.show()
```

```python
# Punto 4) χ2(5)
m = 5*n

etaxx = np.zeros(m)

for i in range (int(m/2)):
    etax1 = np.random.rand() # eta ~ U (0,1)
    etax2 = np.random.rand() # eta ~ U (0,1)
    
    etaxx [i] = np.sqrt(-2*np.log(etax1)) * np.cos(2*np.pi * etax2)
    etaxx [i+int(m/2)] = np.sqrt(-2*np.log(etax1)) * np.sin(2*np.pi * etax2)

n_strano = np.zeros(n)
for i in range (n):
    n_strano [i] = (etaxx[(i*5)]**2) + (etaxx[1+(i*5)]**2) + (etaxx[2+(i*5)]**2) + (etaxx[3+(i*5)]**2) + (etaxx[4+(i*5)]**2)

# Istogramma
x4_graph = np.linspace(min(n_strano), max(n_strano), n)
y4_graph = chi2.pdf(x4_graph, 5)
plt.hist(n_strano, 100, density = True, edgecolor = "black")
plt.title ("Distribuzione chi²")
plt.plot (x4_graph, y4_graph)
plt.show()

print(" ")
# Grafico quantile quantile
probplot(n_strano, dist = chi2(5), plot = plt)
plt.show()
```
