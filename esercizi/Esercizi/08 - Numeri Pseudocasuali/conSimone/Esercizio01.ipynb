```python
# Scrivere un algoritmo per generare numeri pseudo casuali con distribuzione
# multinomiale

import numpy as np
p = np.array([1/4, 1/2, 1/4])

# Generare 2000 di questi numeri e costruire un istogramma verificando l'accordo
# con la distribuzione teorica di probabilità
```

```python
def rand_multi (p):
    F = np.cumsum (p)
    xi = np.random.rand()
    
    if xi < F[0]:
        return 1
    if xi < F[1]:
        return 2
    if xi < F[2]:
        return 3
```

```python
# Punto 1) # Generare 2000 di questi numeri e costruire un istogramma verificando
# l'accordo con la distribuzione teorica di probabilità
n = 2000
X = np.zeros(n)
for i in range (n):
    X[i] = rand_multi(p)

#print ("Ecco X:", X)

m = p.size
F = np.zeros(m)
for i in range (m):
    F[i] = np.sum(X == i+1)

print ("Ecco F:", F)

import matplotlib.pyplot as plt

x_graph = [1, 2, 3]
y_graph = F / n
plt.bar (x_graph, y_graph)
plt.title("Istogramma")
plt.xlabel("Probabilità")
plt.ylabel("Numero occorrenze")
plt.show()
```
