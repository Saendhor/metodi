```python
#Il peso corporeo e la pressione sistolica del sangue di 26 individui maschi selezionati in modo casuale nella fascia d'eta 
#che va da 25 a 30 anni sono mostrati in tabella.
#Assumiamo che il peso e la pressione sanguigna siano normalmente distribuiti.
```

```python
#1: si determini la retta di regressione
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

DATA = np.loadtxt("DATA_reg_lin.dat")
x = DATA[:,1]
y = DATA[:,2]
n = x.size #è uguale a quella di y

print("x: ", x)
print("y: ", y)

#prima del grafico dobbiamo calcolare i coefficenti di regressione beta0 e beta1 --> y = beta0 + beta1*x

#x e y seganto --> sono le medie 
x_media = np.mean(x)
y_media = np.mean(y)

#sigma xy
sigma_xy = np.sum((x-x_media) * (y-y_media))/n
print("sigma_xy: ", sigma_xy)

#varianza x 
var_x = np.sum((x-x_media)**2)/n
print("varianza_x: ", var_x)

#ora calcoliamo b0 e b1
b0 = y_media - ((sigma_xy/var_x)*x_media)
b1 = sigma_xy / var_x

#equazione retta: y = b0 + b1*x
xx = np.linspace(140., 240., 1000) #il primo e il secondo parametro sono il più piccolo e il piùgrande valore dell'asse x (nei dati)
yy = b0 + b1*xx 

plt.plot(x, y, '*')
plt.plot(xx,yy)
plt.xlabel("peso corporeo")
plt.ylabel("pressione sistolica")
plt.title("Retta di regressione")
plt.show()



```

```python
#2: si calcolino gli intervalli di confidenza per i coefficenti di regressione
```

```python
#calcolaimo i valori stimati e i residui
y_stimati = b0 + b1*x
r_residui = y - y_stimati

#mostriamo il grafico dei residui
plt.title("grafico dei residui")
plt.xlabel("peso corporeo")
plt.ylabel("residui")
plt.plot(x,r_residui,'*')
plt.show()
```

```python
from math import sqrt
from scipy.stats import t

#adesso calcoliamo s
s2 = np.sum(r_residui**2) / (n-2)
print("s2: ", s2)
s = sqrt(s2)
alpha = 0.05
T = t.ppf(1-alpha/2, n-2)
dev_standx = sqrt(var_x)

#ricaviamo gli intervalli di confidenza
b0S = b0-s*sqrt(1/n+x_media**2/(n*var_x)) * T
b0D = b0+s*sqrt(1/n+x_media**2/(n*var_x)) * T

b1S = b1-s/(dev_standx*sqrt(n))*T
b1D = b1+s/(dev_standx*sqrt(n))*T

print(
    "Gli intervalli di confidenza sono: ",
    "[",
    b0S,b0D,
    "]",
    "[",
    b1S,b1D,
    "]"
)
```

```python
#3: si testi la significatività della regeressione usando alpha = 0.05
```

```python
#usiamo il test di ipotesi

#calcoliamo la variabile
X = np.abs(sqrt(n)*b1/s*dev_standx)
print("X: ", X)
print("T: ",T)

if X >= T:
    print("L'ipotesi nulla si può rigettare.")
else:
    print("L'ipotesi nulla non ouò essere rigettata.")
```

```python
#4: si calcoli il coefficente di determinazione
```

```python
#calcolaimo la varianza degli scarti e la varianza dei valori stimati
var_scarti = np.sum(r_residui**2)/n
var_stimati = sigma_xy**2 / var_x

print("varianza degli scarti: ", var_scarti)
print("varianza dei valori stimati: ", var_stimati)

var_y = var_scarti + var_stimati
print("varianza_y: ", var_y)

#ricaviamo il coefficente di determinazione
R = 1 - var_scarti / var_y 
print("Il coefficente di determinazione è: ", R)
```
