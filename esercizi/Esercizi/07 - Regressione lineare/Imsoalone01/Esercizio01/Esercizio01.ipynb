```python
# Il peso corporeo e la pressione sistolica del sangue di 26 individui maschi selezionati in modo
# casuale nella fascia d’et`a che va da 25 a 30 anni sono mostrati in tabella. Assumiamo che il
# peso e la pressione sanguigna siano normalmente distribuiti.
```

```python
import numpy as np

#Importo la tabella
DATA = np.loadtxt("Esercizio01_DATA.dat")

# Creo degli array con indice corrispondente al soggetto
# e per ogni soggetto salvo peso e pressione sistolica
# In data la colonna 0 rappresenta i soggetti, la colonna 1 il peso e la colonna 2 la pressione sistolica
x = DATA [: , 1] # Peso
y = DATA [: , 2] # Pressione sistolica

print ("I valori peso sono =>", x)
print ("I valori di pressione sistolica sono =>", y)
```

```python
# Punto 1) Si determini la retta di regressione lineare  SLIDE: 13 ~ 14

# y = b0 + b1 x

x_medio = np.mean (x)
y_medio = np.mean (y)
n = x.size
print ("x medio", x_medio, "y medio", y_medio, "num elementi", n)

sig_xy = np.sum ((x-x_medio) * (y-y_medio))/n
sig_x_quadro = np.sum ((x - x_medio)**2.)/n
print ("sig xy", sig_xy, "sig x²", sig_x_quadro)

b1 = sig_xy / sig_x_quadro
b0 = y_medio - (x_medio * (sig_xy / sig_x_quadro))
print ("I valori di b0 e b1 sono", b0, b1)
```

```python
import matplotlib.pyplot as plt

x_graph = np.linspace (min(x) , max (x), 1000)
y_graph = b0 + b1 * x_graph

print ("Il minimo in x è", min(x), "mentre il massimo in x è", max(x))

plt.plot (x, y, '*')
plt.plot (x_graph, y_graph)
plt.xlabel ('Peso corporeo')
plt.ylabel ('Pressione sistolica')
plt.show ()
```

```python
# Punto 2) Si calcolino gli intervalli di confidenza per i coefficienti di regressione.
from scipy.stats import t

# Calcolo i residui con annessi i valori stimati
# I residui costituiscono la parte di variabilità
# che non è spiegata dal modello
y_hat = b0 + b1*x
r = y - y_hat

print ("Il valore di r è", r)
print (" ")

plt.plot (x, r, '*')
plt.xlabel ("Peso corporeo")
plt.ylabel ("Residuo")
plt.show ()

# Calcolo lo stimatore non distorto
s_quadro = np.sum(r**2.)/(n-2)
s = np.sqrt(s_quadro)

print (" ")
print ("Il valore di s è", s)

# determino l'intervallo di confidenza di livello 1-alpha
alpha = 0.05
T = t.ppf (1. - alpha/2, n-2)

b0_left = b0 - s * np.sqrt(1/n + x_medio**2/(n*sig_x_quadro)) * T
b0_right = b0 + s * np.sqrt(1/n + x_medio**2/(n*sig_x_quadro)) * T

b1_left = b1 - s / np.sqrt(n * sig_x_quadro) * T
b1_right = b1 + s / np.sqrt(n * sig_x_quadro) * T

print (" ")
print ("L'intervallo di confidenza di b0 =", b0 ,"[", b0_left, ",", b0_right, "]")
print ("L'intervallo di confidenza di b1 =", b1 ,"[", b1_left, ",", b1_right, "]")
```

```python
# Punto 3) Si testi la significatività della regressione usando α = 0.05

# Test di significatività
# H0 : beta_1 = 0
# H1 : beta_1 <> 0

h1 = np.sqrt(n) * b1 / s * np.sqrt(sig_x_quadro)
print ("Il valore del test è ", h1)
print ("Il valore di T è", T)
# Si rigetta l'ipotesi nulla in favore di quella alternativa
```

```python
# Punto 4) Si calcoli il coefficiente di determinazione

sig_y_quadro = np.sum ((y - y_medio)**2.)/n

R_quadro = sig_xy**2. / (sig_x_quadro * sig_y_quadro)

print ("Il coefficiente di determinazione R² è", R_quadro)
# IL modello di regressione non sembra essere il più adeguato (deve essere prossimo a 1)
```
